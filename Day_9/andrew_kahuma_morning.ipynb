```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('UceResultsBySchool2011-2016.csv')
df
```

```python
# Sample out some of the data | Dataset before cleaning
df.head()
# Check the shape | Helps to determine No of Rows 
df.shape
```

```python
# Preparation and Preprocessing or data cleaning
df.columns
# Removing whitespaces
df.columns = df.columns.to_series().apply(lambda x: x.strip())
df.columns
```

```python
# Replace column names with underscore
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
# Check for the information of the dataset
df.info()
print("================")
# Check for the duplicates
for is_dup in df.duplicated():
    if is_dup:
        print('Duplicates Identified and dropped!')
        df.drop_duplicates(inplace=True)
        break
    else:
        pass
print('No duplicates found!') 
df[df.duplicated()==True]   
```

```python
# Check for the number of rows with null values
null_rows = df[df.isna().any(axis=1)]
before_compution = len(null_rows.axes[0])
print('Number of rows that are null values:',str(before_compution))
```

```python
# Remove rows without school
df.dropna(thresh=None, subset=['SCHOOL'], inplace=True)
```
