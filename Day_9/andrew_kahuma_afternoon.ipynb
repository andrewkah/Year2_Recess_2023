### DATA ANALYSIS WITH PANDAS
### SERIES

```python
import pandas as pd
```

```python
# Comparing Pandas series
sample_1 = [4, 65, 436, 3, 9]
sample_2 = [7,0,3,897,9]
compare = sample_1 == sample_2
print(compare)
```

```python
# add, subtract, multiply and divide two Pandas series
sample_1 = [2,4,6,8,14]
sample_2 = [1,3,5,7,9]
addition = sample_1 + sample_2
subtraction = sample_1 - sample_2
multiplication = sample_1 * sample_2
division = sample_1 / sample_2
print(f'Addition: {addition}')
print(f"Subtraction: {subtraction}")
print(f'Multiplication: {multiplication}')
print(f'Division: {division}')
```

```python
# converting a dictionary to pandas series
dictionary1 = {'Josh':24, 'Sam':36, 'Peace':19, 'Charles':65, 'Tom':44}
sereies = pd.Series(dictionary1)
print(sereies)
```

```python
# Convert a series to an array
import numpy as np
series = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])
array = np.array(series)
print(array)
```

```python
# Display the most frequent value in the series
data = pd.read_csv('BSSE 2301 EXERCISE 2/AfricaCupofNationsMatches.csv')
df_series = pd.Series(data['HomeTeamGoals'])
most_frequent = df_series.value_counts().idxmax()
df_series[~(df_series == most_frequent)] = 'Other'

print(df_series)
```

```python
import pandas as pd
import numpy as np
```

```python
# Reading the csv file
data_file = pd.read_csv('BSSE 2301 EXERCISE 2/AfricaCupofNationsMatches.csv')
```

```python
# Getting the first 7 rows of the dataframe
df = pd.DataFrame(data_file)
first_seven_rows = df.head(7)
first_seven_rows
```

```python
# Select the 'HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals' columns
selected_columns = df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]
print(selected_columns)
```

```python
# select the rows where Egypt appears
egypt_rows = df[(df['HomeTeam'] == 'Egypt')| (df['AwayTeam'] == 'Egypt')]
egypt_rows
```

```python
# count the number of rows and columns of the data
rows, columns = df.shape
print(f'Number of rows: {rows}')
print(f'Number of columns: {columns}')
```

```python
# selecting the rows where 'Attendance' is missing
missing_attendance = df[df['Attendance'].isnull()]
missing_attendance
```

```python
# selecting rows where 'HomeTeamGoals' are between 3 and 6
between_3_and_6 = df[(df['HomeTeamGoals']>=3) & (df['HomeTeamGoals']<=6)]
between_3_and_6
```

```python
# change the 'AwayTeamGoals' in the 3rd row to 10
df.at[2, 'AwayTeamGoals'] = 10
print(df.head())
```

```python
# sort the dataframe, 'HomeTeam' in ascending order and 'HomeTeamGoals' in descending order
sorted_df = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])
sorted_df
```

```python
# Get the list from DataFrame column headers
column_headers = list(df.columns)
column_headers
```
